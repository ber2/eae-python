En este notebook vamos a:
1. Calcular medidas de tendencia central y dispersión
2. Detectar outliers con el IQR
3. Hacer análisis de frecuencias
4. Calcular e interpretar correlaciones

```python
import pandas as pd
import numpy as np
import plotly.express as px
df = pd.read_csv("https://raw.githubusercontent.com/ber2/eae-python/main/data/Houses_Barcelona_samp.csv")
df.describe()
```

## Parte 1: Tendencia central

Media, mediana y moda del precio.

```python
media = df["price"].mean()
mediana = df["price"].median()
moda = df["price"].mode()[0]
print(f"Media:   {media:>10,.0f} €")
print(f"Mediana: {mediana:>10,.0f} €")
print(f"Moda:    {moda:>10,.0f} €")
print("\nLa media > mediana → distribución asimétrica a la derecha")
```

```python
fig = px.histogram(df, x="price", nbins=50, title="Distribución de precios")
fig.add_vline(x=media, line_dash="dash", line_color="red", annotation_text="Media", )
fig.add_vline(x=mediana, line_dash="dash", line_color="green", annotation_text="Mediana")
fig.show()
```

## Parte 2: Dispersión

Rango, varianza y desviación estándar.

```python
rango = df["price"].max() - df["price"].min()
print(f"Rango:      {rango:>12,.0f} €")
print(f"Varianza:   {df['price'].var():>12,.0f} €²")
print(f"Desv. Est.: {df['price'].std():>12,.0f} €")
```

### Cuartiles, IQR y outliers

```python
Q1, Q3 = df["price"].quantile([0.25, 0.75])
IQR = Q3 - Q1
li, ls = Q1 - 1.5*IQR, Q3 + 1.5*IQR
outliers = df[(df["price"] < li) | (df["price"] > ls)]
print(f"Q1={Q1:,.0f}  Q3={Q3:,.0f}  IQR={IQR:,.0f}")
print(f"Límites outliers: [{li:,.0f}, {ls:,.0f}]")
print(f"Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")
```

```python
px.box(df, y="price", title="Box plot de precios (outliers incluidos)",
       labels={"price":"Precio (€)"})
```

## Parte 3: Análisis de frecuencias

Frecuencias absolutas y relativas de una variable categórica.

```python
conteo = df["neighborhood"].value_counts()
relativo = df["neighborhood"].value_counts(normalize=True)
freq = pd.DataFrame({"frecuencia": conteo, "proporción": relativo.round(3)})
freq
```

```python
conteo_tipo = df["type"].value_counts().reset_index()
conteo_tipo.columns = ["type", "count"]
px.bar(conteo_tipo, x="type", y="count", title="Tipos de propiedad")
```

### Resumen por grupo

```python
resumen = df.groupby("neighborhood")["price"].agg(["mean","median","count"]).round(0)
resumen.sort_values("mean", ascending=False)
```

## Parte 4: Correlación

Relación lineal entre variables numéricas.

```python
r = df["price"].corr(df["sqrmts"])
print(f"Correlación precio ~ metros²: {r:.3f}")
px.scatter(df, x="sqrmts", y="price", trendline="ols",
           title=f"Precio vs Metros Cuadrados (r = {r:.3f})",
           labels={"sqrmts":"Metros Cuadrados","price":"Precio (€)"})
```

```python
cols = ["price","sqrmts","rooms","bathrooms","floor"]
corr = df[cols].corr()
px.imshow(corr, text_auto=".2f", color_continuous_scale="RdBu_r",
          zmin=-1, zmax=1, title="Matriz de Correlación")
```

**Recordad**: correlación ≠ causalidad. Una correlación fuerte no implica que una variable cause la otra.
