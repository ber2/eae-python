## SoluciÃ³n Ejercicio 1

```python
media = df["bathrooms"].mean()
mediana = df["bathrooms"].median()
moda = df["bathrooms"].mode()[0]

print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.0f}")
print(f"Moda: {moda:.0f}")

# La mediana o moda reflejan mejor el valor tÃ­pico
# Si media â‰ˆ mediana â†’ simÃ©trica
# Si media > mediana â†’ asimÃ©trica derecha

skew = df["bathrooms"].skew()
print(f"\nSkewness: {skew:.2f}")
if skew > 0.5:
    print("DistribuciÃ³n asimÃ©trica a la derecha")
elif skew < -0.5:
    print("DistribuciÃ³n asimÃ©trica a la izquierda")
else:
    print("DistribuciÃ³n aproximadamente simÃ©trica")
```

## SoluciÃ³n Ejercicio 2

```python
# 1. Rango
rango = df["sqrmts"].max() - df["sqrmts"].min()
print(f"Rango: {rango:.0f} mÂ²")

# 2. DesviaciÃ³n estÃ¡ndar
std = df["sqrmts"].std()
print(f"DesviaciÃ³n estÃ¡ndar: {std:.1f} mÂ²")

# 3. Cuartiles e IQR
Q1 = df["sqrmts"].quantile(0.25)
Q3 = df["sqrmts"].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1: {Q1:.0f}, Q3: {Q3:.0f}, IQR: {IQR:.0f}")

# 4. Outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers = df[(df["sqrmts"] < limite_inferior) | (df["sqrmts"] > limite_superior)]
print(f"Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")
```

## SoluciÃ³n Ejercicio 3

```python
# 1 y 2. Correlaciones
corr_rooms = df["price"].corr(df["rooms"])
corr_bathrooms = df["price"].corr(df["bathrooms"])

print(f"CorrelaciÃ³n price-rooms: {corr_rooms:.3f}")
print(f"CorrelaciÃ³n price-bathrooms: {corr_bathrooms:.3f}")

# 3. Â¿CuÃ¡l es mÃ¡s fuerte?
if abs(corr_rooms) > abs(corr_bathrooms):
    print(f"\nrooms tiene correlaciÃ³n mÃ¡s fuerte")
    var_mas_fuerte = "rooms"
    corr_mas_fuerte = corr_rooms
else:
    print(f"\nbathrooms tiene correlaciÃ³n mÃ¡s fuerte")
    var_mas_fuerte = "bathrooms"
    corr_mas_fuerte = corr_bathrooms

# 4. Scatter plot
fig = px.scatter(df, x=var_mas_fuerte, y="price",
                 title=f"Precio vs {var_mas_fuerte} (r = {corr_mas_fuerte:.3f})",
                 trendline="ols")
fig.show()
```

## SoluciÃ³n Ejercicio 4

```python
total = len(df)

# 1. P(Ãtico)
prob_atico = len(df[df["type"] == "Ãtico"]) / total
print(f"P(Ãtico) = {prob_atico:.3f}")

# 2. P(precio > 500k)
prob_caro = len(df[df["price"] > 500000]) / total
print(f"P(precio > 500k) = {prob_caro:.3f}")

# 3. P(Ãtico Y precio > 500k)
prob_ambos = len(df[(df["type"] == "Ãtico") & (df["price"] > 500000)]) / total
print(f"P(Ãtico Y precio > 500k) = {prob_ambos:.3f}")

# 4. Â¿Independientes?
prob_si_indep = prob_atico * prob_caro
print(f"\nSi fueran independientes: P(A) Ã— P(B) = {prob_si_indep:.3f}")
print(f"Valor real: {prob_ambos:.3f}")

if abs(prob_ambos - prob_si_indep) < 0.01:
    print("âœ“ Aproximadamente independientes")
else:
    print("âœ— NO son independientes")
    if prob_ambos > prob_si_indep:
        print("  â†’ Ser Ãtico aumenta la probabilidad de ser caro")
    else:
        print("  â†’ Ser Ãtico disminuye la probabilidad de ser caro")
```

## SoluciÃ³n Ejercicio 5

```python
# 1. ParÃ¡metros
mu = df["price"].mean()
sigma = df["price"].std()
print(f"Î¼ = {mu:,.0f}, Ïƒ = {sigma:,.0f}")

# 2. Intervalos
int_1sigma = (mu - sigma, mu + sigma)
int_2sigma = (mu - 2*sigma, mu + 2*sigma)
int_3sigma = (mu - 3*sigma, mu + 3*sigma)

print(f"\nÎ¼ Â± 1Ïƒ: [{int_1sigma[0]:,.0f}, {int_1sigma[1]:,.0f}]")
print(f"Î¼ Â± 2Ïƒ: [{int_2sigma[0]:,.0f}, {int_2sigma[1]:,.0f}]")
print(f"Î¼ Â± 3Ïƒ: [{int_3sigma[0]:,.0f}, {int_3sigma[1]:,.0f}]")

# 3. % real
pct_1sigma = ((df["price"] >= int_1sigma[0]) & (df["price"] <= int_1sigma[1])).sum() / len(df)
pct_2sigma = ((df["price"] >= int_2sigma[0]) & (df["price"] <= int_2sigma[1])).sum() / len(df)
pct_3sigma = ((df["price"] >= int_3sigma[0]) & (df["price"] <= int_3sigma[1])).sum() / len(df)

print(f"\n% en Î¼ Â± 1Ïƒ: {pct_1sigma*100:.1f}% (esperado: 68%)")
print(f"% en Î¼ Â± 2Ïƒ: {pct_2sigma*100:.1f}% (esperado: 95%)")
print(f"% en Î¼ Â± 3Ïƒ: {pct_3sigma*100:.1f}% (esperado: 99.7%)")

# 4. Â¿Normal?
print("\nÂ¿Es normal? NO - Los porcentajes difieren de 68-95-99.7")
print("Los precios tienen skew positivo (asimÃ©tricos a la derecha)")
print(f"Skewness = {df['price'].skew():.2f} (deberÃ­a ser ~0 para normal)")
```

## SoluciÃ³n Ejercicio Bonus

```python
# 1. Z-scores
df["price_zscore"] = (df["price"] - df["price"].mean()) / df["price"].std()

# 2. Top 5 extremas
df["price_zscore_abs"] = df["price_zscore"].abs()
extremas = df.nlargest(5, "price_zscore_abs")

print("5 propiedades mÃ¡s extremas:")
print(extremas[["neighborhood", "type", "price", "sqrmts", "rooms", "price_zscore"]])

# 3. Â¿QuÃ© tienen en comÃºn?
print("\nÂ¿QuÃ© tienen en comÃºn?")
print("- Todas tienen z-score muy alto (>2 o >3)")
print("- Son propiedades MUY caras (precio muy por encima de la media)")
print("- Probablemente en barrios caros y/o con muchos metros")
```
