```python
# Solución 1
p3 = (df["rooms"] >= 3).mean()
gr = df[df["neighborhood"] == "Gràcia"]
p3_gr = (gr["rooms"] >= 3).mean()
print(f"P(3+ hab) = {p3:.3f}")
print(f"P(3+ hab | Gràcia) = {p3_gr:.3f}")
```

```python
# Solución 2
d = stats.binom(n=20, p=0.1)
print("P(exactamente 3) =", round(d.pmf(3), 3))
print("P(5 o más)       =", round(1 - d.cdf(4), 3))
```

```python
# Solución 3
mu, sigma = df["sqrmts"].mean(), df["sqrmts"].std()
print(f"μ = {mu:.1f} m², σ = {sigma:.1f} m²")
print("P(sqrmts ≤ 100) teórica =", round(stats.norm(mu, sigma).cdf(100), 3))
real = df["sqrmts"].between(70, 100).mean()
print(f"% real entre 70-100 m²: {real:.1%}")
```
