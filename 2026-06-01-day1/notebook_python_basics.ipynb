## 1. Variables y Tipos de Datos

Las **variables** son contenedores que almacenan información. En Python, no necesitas declarar el tipo de variable - Python lo deduce automáticamente.

```python
# Creamos diferentes tipos de variables
edad = 28                    # int (número entero)
altura = 1.75                # float (número decimal)
nombre = "Ana"               # str (string/texto)
es_estudiante = True         # bool (booleano: True o False)

# Imprimir variables
print("Edad:", edad)
print("Altura:", altura)
print("Nombre:", nombre)
print("¿Es estudiante?", es_estudiante)
```

```python
# Verificar el tipo de una variable con type()
print(type(edad))           # <class 'int'>
print(type(altura))         # <class 'float'>
print(type(nombre))         # <class 'str'>
print(type(es_estudiante))  # <class 'bool'>
```

## 2. Operadores

### Operadores Aritméticos

```python
# Operaciones básicas
a = 10
b = 3

print("Suma:", a + b)              # 13
print("Resta:", a - b)             # 7
print("Multiplicación:", a * b)    # 30
print("División:", a / b)          # 3.333...
print("División entera:", a // b)  # 3
print("Módulo (resto):", a % b)    # 1
print("Potencia:", a ** b)         # 1000 (10 elevado a 3)
```

```python
# Orden de operaciones (igual que en matemáticas)
resultado1 = 2 + 3 * 4      # Primero multiplica, luego suma: 14
resultado2 = (2 + 3) * 4    # Paréntesis primero: 20

print("Sin paréntesis:", resultado1)
print("Con paréntesis:", resultado2)
```

### Operadores de Comparación

```python
# Comparaciones devuelven True o False
x = 10
y = 5

print("x == y:", x == y)   # ¿Son iguales? False
print("x != y:", x != y)   # ¿Son diferentes? True
print("x > y:", x > y)     # ¿x mayor que y? True
print("x < y:", x < y)     # ¿x menor que y? False
print("x >= 10:", x >= 10) # ¿x mayor o igual a 10? True
print("y <= 5:", y <= 5)   # ¿y menor o igual a 5? True
```

### Operadores Lógicos

```python
# Combinar condiciones con and, or, not
edad = 25
tiene_licencia = True

# AND: ambas condiciones deben ser True
puede_conducir = (edad >= 18) and tiene_licencia
print("¿Puede conducir?", puede_conducir)

# OR: al menos una condición debe ser True
dia = "sábado"
es_fin_de_semana = (dia == "sábado") or (dia == "domingo")
print("¿Es fin de semana?", es_fin_de_semana)

# NOT: invierte el valor
es_dia_laboral = not es_fin_de_semana
print("¿Es día laboral?", es_dia_laboral)
```

## 3. Strings y F-strings

Los **strings** (cadenas de texto) son fundamentales para trabajar con datos.

```python
# Crear strings
nombre = "Carlos"
apellido = 'García'  # Comillas simples o dobles

# Concatenar (unir) strings
nombre_completo = nombre + " " + apellido
print(nombre_completo)

# Repetir strings
separador = "-" * 30
print(separador)

# Longitud de un string
print("Longitud del nombre:", len(nombre))
```

```python
# Métodos útiles de strings
texto = "  Python es Genial  "

print("Original:", texto)
print("Mayúsculas:", texto.upper())
print("Minúsculas:", texto.lower())
print("Sin espacios:", texto.strip())
print("Reemplazar:", texto.replace("Genial", "Increíble"))
```

### F-strings: La forma moderna de formatear texto

```python
# F-strings permiten insertar variables dentro de strings
nombre = "Ana"
edad = 28
altura = 1.75
sueldo = 45000.50

# Forma básica
mensaje = f"Me llamo {nombre} y tengo {edad} años"
print(mensaje)

# Con formato de números
print(f"Altura: {altura:.2f} metros")       # 2 decimales
print(f"Sueldo: €{sueldo:,.2f}")             # Con separador de miles
print(f"Porcentaje: {0.156:.1%}")            # Como porcentaje

# Con expresiones
print(f"El doble de mi edad es {edad * 2}")
print(f"Mi nombre en mayúsculas: {nombre.upper()}")
```

## 4. Control de Flujo: if/elif/else

Permite que el código tome decisiones basadas en condiciones.

```python
# if básico
edad = 20

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

```python
# if/elif/else para múltiples condiciones
nota = 75

if nota >= 90:
    calificacion = "A - Excelente"
elif nota >= 80:
    calificacion = "B - Muy bien"
elif nota >= 70:
    calificacion = "C - Bien"
elif nota >= 60:
    calificacion = "D - Suficiente"
else:
    calificacion = "F - Insuficiente"

print(f"Tu calificación: {calificacion}")
```

```python
# Ejemplo de negocio: Descuentos por volumen
cantidad = 150
precio_unitario = 10

if cantidad >= 100:
    descuento = 0.20  # 20% de descuento
    print("Descuento aplicado: 20%")
elif cantidad >= 50:
    descuento = 0.10  # 10% de descuento
    print("Descuento aplicado: 10%")
else:
    descuento = 0     # Sin descuento
    print("Sin descuento")

precio_bruto = cantidad * precio_unitario
ahorro = precio_bruto * descuento
precio_final = precio_bruto - ahorro

print(f"Precio sin descuento: €{precio_bruto:,.2f}")
print(f"Ahorro: €{ahorro:,.2f}")
print(f"Precio final: €{precio_final:,.2f}")
```

## 5. Bucles

Los bucles permiten repetir código múltiples veces.

### Bucle for

```python
# Iterar sobre una lista
departamentos = ["Marketing", "Ventas", "IT", "Finanzas"]

for departamento in departamentos:
    print(f"Departamento: {departamento}")
```

```python
# Iterar sobre un rango de números
print("Contando del 1 al 5:")
for i in range(1, 6):
    print(f"Número {i}")

print("\nContando del 0 al 4:")
for i in range(5):  # Empieza en 0 por defecto
    print(f"Número {i}")
```

```python
# Ejemplo de negocio: Calcular total de ventas
ventas_diarias = [1200, 1500, 1800, 1350, 2000, 2200, 1900]
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom']

total = 0
for i in range(len(ventas_diarias)):
    venta = ventas_diarias[i]
    dia = dias[i]
    print(f"{dia}: €{venta:,}")
    total += venta  # Equivale a: total = total + venta

promedio = total / len(ventas_diarias)
print(f"\nTotal semanal: €{total:,}")
print(f"Promedio diario: €{promedio:,.2f}")
```

### Bucle while

```python
# Contador simple con while
contador = 1

while contador <= 5:
    print(f"Vuelta número {contador}")
    contador += 1

print("Bucle terminado")
```

```python
# Ejemplo: Doblar inversión hasta alcanzar meta
capital_inicial = 1000
meta = 5000
tasa_anual = 0.10  # 10% anual
años = 0

capital_actual = capital_inicial

while capital_actual < meta:
    capital_actual = capital_actual * (1 + tasa_anual)
    años += 1
    print(f"Año {años}: €{capital_actual:,.2f}")

print(f"\n¡Meta alcanzada en {años} años!")
print(f"Capital final: €{capital_actual:,.2f}")
```

## 6. Funciones

Las funciones son bloques de código reutilizables que realizan una tarea específica.

```python
# Función simple
def saludar(nombre):
    """Saluda a una persona por su nombre"""
    mensaje = f"Hola, {nombre}! Bienvenido/a"
    return mensaje

# Usar la función
saludo1 = saludar("Ana")
saludo2 = saludar("Carlos")

print(saludo1)
print(saludo2)
```

```python
# Función con múltiples parámetros
def calcular_precio_final(precio, descuento_porcentaje, iva=0.21):
    """
    Calcula el precio final aplicando descuento e IVA
    
    Args:
        precio: precio base del producto
        descuento_porcentaje: descuento a aplicar (0-100)
        iva: tasa de IVA (default 0.21 = 21%)
    
    Returns:
        precio final con descuento e IVA aplicados
    """
    descuento = precio * (descuento_porcentaje / 100)
    precio_con_descuento = precio - descuento
    precio_final = precio_con_descuento * (1 + iva)
    
    return precio_final

# Usar la función
precio1 = calcular_precio_final(100, 10)         # IVA por defecto (21%)
precio2 = calcular_precio_final(100, 10, 0.10)   # IVA del 10%

print(f"Precio con IVA 21%: €{precio1:.2f}")
print(f"Precio con IVA 10%: €{precio2:.2f}")
```

```python
# Función sin return (hace algo pero no devuelve valor)
def imprimir_reporte(ventas, mes):
    """Imprime un reporte formateado de ventas"""
    print("=" * 40)
    print(f"REPORTE DE VENTAS - {mes.upper()}")
    print("=" * 40)
    print(f"Total de ventas: €{sum(ventas):,.2f}")
    print(f"Venta promedio: €{sum(ventas)/len(ventas):,.2f}")
    print(f"Venta máxima: €{max(ventas):,.2f}")
    print(f"Venta mínima: €{min(ventas):,.2f}")
    print("=" * 40)

# Usar la función
ventas_enero = [1200, 1500, 1800, 1350, 2000]
imprimir_reporte(ventas_enero, "Enero")
```
