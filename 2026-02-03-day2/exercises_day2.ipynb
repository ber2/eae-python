## Soluci√≥n Ejercicio 1.1

```python
metros = [85, 120, 65, 90, 110, 75]

# Media
media = sum(metros) / len(metros)
print(f"Media: {media:.1f} m¬≤")

# M√°ximo y m√≠nimo
print(f"M√°ximo: {max(metros)} m¬≤")
print(f"M√≠nimo: {min(metros)} m¬≤")

# Filtrar >80m¬≤
grandes = [m for m in metros if m > 80]
print(f"Propiedades >80m¬≤: {grandes}")
```

## Soluci√≥n Ejercicio 1.2

```python
# Crear diccionario
propiedad = {
    "barrio": "Sarri√†-Sant Gervasi",
    "precio": 520000,
    "habitaciones": 4,
    "metros": 105
}

# Precio por m¬≤
precio_m2 = propiedad["precio"] / propiedad["metros"]
print(f"Precio/m¬≤: {precio_m2:.0f}‚Ç¨/m¬≤")

# A√±adir parking
propiedad["parking"] = True

# Imprimir resumen
print(f"Piso de {propiedad['habitaciones']} habitaciones en {propiedad['barrio']}, {propiedad['precio']}‚Ç¨")
```

## Soluci√≥n Ejercicio 1.3

```python
# Crear lista de propiedades
propiedades = [
    {"barrio": "Gr√†cia", "precio": 280000, "habitaciones": 2},
    {"barrio": "Eixample", "precio": 450000, "habitaciones": 3},
    {"barrio": "Sants", "precio": 195000, "habitaciones": 1}
]

# Filtrar <300k
baratas = [p for p in propiedades if p["precio"] < 300000]
print(f"Propiedades <300k: {len(baratas)}")
for p in baratas:
    print(f"  - {p['barrio']}: {p['precio']}‚Ç¨")
```

## Soluci√≥n Ejercicio 2.1

```python
import pandas as pd

url = "https://raw.githubusercontent.com/ber2/2026-intro-python-ml/main/data/Houses_Barcelona_samp.csv"
df = pd.read_csv(url)

# Dimensiones
print(f"Dimensiones: {df.shape}")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")

# Tipos de datos
print("\nInfo del DataFrame:")
df.info()

# Valores faltantes
print("\nValores faltantes por columna:")
faltantes = df.isnull().sum()
print(faltantes[faltantes > 0])
```

## Soluci√≥n Ejercicio 2.2

```python
# Estad√≠sticas descriptivas
print(df.describe())

# Precio medio
precio_medio = df["price"].mean()
print(f"\nPrecio medio: {precio_medio:.0f}‚Ç¨")

# Metros m√°ximos
metros_max = df["sqrmts"].max()
print(f"Metros de la m√°s grande: {metros_max} m¬≤")

# Valores sospechosos: Observar rooms.max() y sqrmts.max() - pueden tener valores imposibles
```

## Soluci√≥n Ejercicio 3.1
